```python
def make_blank_plot(ax):
    """Use for suploting, not to have the plot space when no need"""
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.set_xticks([])
    ax.set_yticks([])
```

```python
h = 7
l = 2

f, ax = plt.subplots(h, l, figsize=(20,30))
for i in range(h):
    for j in range(l):
        if i*l+j < data.shape[1]:
            sns.kdeplot(data.iloc[:, i*l+j], ax=ax[i, j])
            ax[i, j].set_title(f'Distribution of {data.columns[i*l+j]}')
        else:
            make_blank_plot(ax[i, j])
plt.tight_layout()
plt.show()
```

```python
h = 7
l = 2

f, ax = plt.subplots(h, l, figsize=(20,30))
for i in range(h):
    for j in range(l):
        if i*l+j < data.shape[1]:
            sns.boxplot(data.iloc[:, i*l+j], ax=ax[i, j])
            ax[i, j].set_title(f'Distribution of {data.columns[i*l+j]}')
        else:
            make_blank_plot(ax[i, j])
plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split

target = 'Bankrupt?'
X = data.copy()
del X[target]
y = data[target]

X_train, X_test, y_train, y_test = train_test_split(X, y,
                     test_size=.2, stratify=y)
```

```python
X_train.columns
```

```python
def clip_data(feat_name):
    Q1 = X_train[feat_name].quantile(0.25)
    Q3 = X_train[feat_name].quantile(0.75)
    
    IQR = Q3 - Q1
    
    lower_bound_clip = Q1 - 1.5 * IQR
    upper_bound_clip = Q3 + 1.5 * IQR
    
    X_train[feat_name] = X_train[feat_name].clip(lower_bound_clip, upper_bound_clip)
    X_test[feat_name] = X_test[feat_name].clip(lower_bound_clip, upper_bound_clip)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import ShuffleSplit
from sklearn.ensemble import GradientBoostingClassifier

params_dict_new = {'loss': ['log_loss', 'exponential'],
                   'learning_rate': [0.1, 0.01],
                   'n_estimators': [2, 4, 6],
                   'criterion': ['friedman_mse', 'squared_error', 'squared_error'],
                   'min_samples_leaf': [2, 5, 10],
                  }
metric = 'neg_log_loss'

print(f'Tuning hyper-parameters with Randomized Search for {metric}')

clf_gbc = RandomizedSearchCV(GradientBoostingClassifier(),
                             params_dict_new,
                             scoring=metric,
                             cv=ShuffleSplit(n_splits=5),
                             random_state=0,
                             verbose=10,
                             n_iter=50,
                             n_jobs=-1)
clf_gbc.fit(X_train, y_train)
```
